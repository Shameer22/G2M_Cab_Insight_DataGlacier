```python
from datetime import datetime, timedelta
def to_date_format(n):
    date_str =(datetime(1899,12,30) + timedelta(n-1)).strftime("%d-%m-%Y")
    date_date =  datetime.strptime(date_str, "%d-%m-%Y")
    return date_date
```

```python
CabData['Date_of_Travel'] = CabData['Date_of_Travel'].apply(lambda x:to_date_format(x))
```

```python
CabData.head()
```

```python
months=[]
years =[]
for i in range(len(CabData['Date_of_Travel'])):
    months.append(CabData['Date_of_Travel'][i].month)
    years.append(CabData['Date_of_Travel'][i].year)
CabData['Month'] = months
CabData['Year'] = years

CabData.drop('Date_of_Travel', axis=1, inplace=True)
```

##Hypothesis Test

###For Pink Cab

```python
a = data[(data.Gender=='Female')&(data.Company=='Pink Cab')].groupby('Transaction_ID').Margins.mean()
b = data[(data.Gender=='Male')&(data.Company=='Pink Cab')].groupby('Transaction_ID').Margins.mean()
print(a.shape[0],b.shape[0])

from scipy import stats
_, p_value = stats.ttest_ind(a.values,b=b.values,equal_var=True)
if(p_value<0.05):
    print('Alternate hypothesis accepted - There is difference')
else:
    print('Null hypothesis accepted - No difference')
    
print('P value: ', p_value)
```

> **No difference in margin between both gender customers**



###For Yellow Cab

```python
a = data[(data.Gender=='Female')&(data.Company=='Yellow Cab')].groupby('Transaction_ID').Margins.mean()
b = data[(data.Gender=='Male')&(data.Company=='Yellow Cab')].groupby('Transaction_ID').Margins.mean()
print(a.shape[0],b.shape[0])

from scipy import stats
_, p_value = stats.ttest_ind(a.values,b=b.values,equal_var=True)
if(p_value<0.05):
    print('Alternate hypothesis accepted - Statistical difference')
else:
    print('Null hypothesis accepted - No statistical difference')
    
print('P value is ', p_value)
```
